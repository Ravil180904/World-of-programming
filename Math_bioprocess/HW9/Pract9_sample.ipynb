```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
L = 10  # length of the spatial domain
T_values = [1.25, 2.5, 5]  # time instances for plotting
Nx = 100  # number of spatial grid points
dx = L / Nx  # spatial step size
c = 2  # speed constant in the PDE
dt = 0.25 * dx / c  # time step, chosen to satisfy stability condition (0.5)

# Initialize spatial and temporal grids
x = np.arange(0, L + dx, dx)  # Create the spatial grid using np.arange
u_initial = np.cos(x) ** 2  # Initial condition

# Simplified function to perform one time step of the Lax-Friedrichs scheme
def lax_friedrichs_step(u, dx, dt, c):
    u[1:-1] = 0.5 * (u[:-2] + u[2:]) - 0.5 * c * dt / dx * (u[2:] - u[:-2])
    u[0] = 1  # boundary condition at x = 0
    # Neumann boundary (or holding the last value constant) at x = L
    return u

# Solve and plot for each T in T_values
plt.figure(figsize=(10, 6))
for T in T_values:
    u = np.copy(u_initial)  # reset to initial condition for each T
    num_steps = int(T / dt)  # Number of time steps
    for i in range(num_steps):  # time loop with explicit index
        u = lax_friedrichs_step(u, dx, dt, c)
    
    # Plotting the result for the given T
    plt.plot(x, u, label=f'T = {T}')

# Plot formatting
plt.xlabel('x')
plt.ylabel('u(x, T)')
plt.title('Solution of the conservation law using Lax-Friedrichs method')
plt.legend()
plt.grid(True)
plt.show()
```
