1. You work in a medical research center. A new test for $HIV$ has been developed. You are asked to evaluate a new test on whether it performs well based on your knowledge of probability theory. You have been given the results of an experiment on a new test for $HIV$.

Legend:

* $HIV^+$ - number of people infected with $HIV$.

* $HIV^-$ - the number of people NOT infected with $HIV$.

* $Test^+$ - the number of people who tested positive for $HIV$.

* $Test^-$ - the number of people whose test showed a negative result for $HIV$.


||$HIV^+$|$HIV^-$|Total|
|:-:|:-:|:-:|:-:|
|$Test^+$|25  |30  |55  |
|$Test^-$|25  |15  |40  |
|Total   |50  |45  |95  |


Based on the data provided to you, please calculate the following probabilities::

* $P(Test^+)$, and assign the result to the variable ```p_tp```.
* $P(Test^-)$, and assign the result to the variable ```p_tn```.
* $P(Test^+,HIV^+)$, and assign the result to the variable ```p_tp_hp```.
* $P(Test^-,HIV^+)$, and assign the result to the variable ```p_tn_hp```.
* $P(Test^+,HIV^-)$, and assign the result to the variable ```p_tp_hn```.
* $P(Test^-,HIV^-)$, and assign the result to the variable ```p_tn_hn```.

````python
#Начало вашего кода
p_tp=55
p_tn=40
p_tp_hp=25
p_tp_hn=30
p_tn_hp=25
p_tn_hn=15
def testpositive():
    a = p_tp_hp + p_tp_hn
    b = p_tn_hn + p_tn_hp
    total = a + b
    n = a/total
    return n
print(testpositive())

#Конец вашего кода

print(f"{p_tp*100:.2f}%")
print(f"{p_tn*100:.2f}%")
print(f"{p_tp_hp*100:.2f}%")
print(f"{p_tn_hp*100:.2f}%")
print(f"{p_tp_hn*100:.2f}%")
print(f"{p_tn_hn*100:.2f}%")
````
