```python
import numpy as np
```

```python
def f(x):
    return 2 / (1 + np.exp(1-x)) - 1
```

```python
def df(x):
    return 0.5 * (1 + x) * (1- x)
```

```python
w1 = np.array([[-0.2, 0.3, - 0.4],[0.1, -0.3, -0.4]])
w2 = np.array([0.2, 0.3])
```

```python
def forward_pass(inp):
    
    sum_ = np.dot(w1, inp)
    out = np.array([f(elem) for elem in sum_])
    
    sum_2 = np.dot(w2, out)
    y = f(sum_2)
    
    return y, out
```
