# Vectors and matrices

## Ex 1. 
Given two vectors:
$$x = \begin{pmatrix} 2 \\ -1 \\ 3 \end{pmatrix}, \quad y = \begin{pmatrix} 5 \\ 0 \\ 4 \end{pmatrix}$$

Compute:
- $x + y$
- $3x - 2y$
- The dot product $x^Ty$
- The outer product $xy^T$

## Vector Calculations

Given two vectors:
\[ x = \begin{pmatrix} 2 \\ -1 \\ 3 \end{pmatrix}, \quad y = \begin{pmatrix} 5 \\ 0 \\ 4 \end{pmatrix} \]

### Sum of Vectors \( x + y \)
\[ 
x + y = \begin{pmatrix} 2 \\ -1 \\ 3 \end{pmatrix} + \begin{pmatrix} 5 \\ 0 \\ 4 \end{pmatrix} = \begin{pmatrix} 2 + 5 \\ -1 + 0 \\ 3 + 4 \end{pmatrix} = \begin{pmatrix} 7 \\ -1 \\ 7 \end{pmatrix} 
\]

### Scalar Multiplication and Difference \( 3x - 2y \)
\[ 
3x = 3 \begin{pmatrix} 2 \\ -1 \\ 3 \end{pmatrix} = \begin{pmatrix} 6 \\ -3 \\ 9 \end{pmatrix} 
\]
\[ 
2y = 2 \begin{pmatrix} 5 \\ 0 \\ 4 \end{pmatrix} = \begin{pmatrix} 10 \\ 0 \\ 8 \end{pmatrix} 
\]
\[ 
3x - 2y = \begin{pmatrix} 6 \\ -3 \\ 9 \end{pmatrix} - \begin{pmatrix} 10 \\ 0 \\ 8 \end{pmatrix} = \begin{pmatrix} 6 - 10 \\ -3 - 0 \\ 9 - 8 \end{pmatrix} = \begin{pmatrix} -4 \\ -3 \\ 1 \end{pmatrix} 
\]

### Dot Product \( x^T y \)
\[ 
x^T y = \begin{pmatrix} 2 \\ -1 \\ 3 \end{pmatrix}^T \begin{pmatrix} 5 \\ 0 \\ 4 \end{pmatrix} = 2 \cdot 5 + (-1) \cdot 0 + 3 \cdot 4 = 10 + 0 + 12 = 22 
\]

### Outer Product \( xy^T \)
\[ 
xy^T = \begin{pmatrix} 2 \\ -1 \\ 3 \end{pmatrix} \begin{pmatrix} 5 & 0 & 4 \end{pmatrix} = \begin{pmatrix} 2 \cdot 5 & 2 \cdot 0 & 2 \cdot 4 \\ -1 \cdot 5 & -1 \cdot 0 & -1 \cdot 4 \\ 3 \cdot 5 & 3 \cdot 0 & 3 \cdot 4 \end{pmatrix} = \begin{pmatrix} 10 & 0 & 8 \\ -5 & 0 & -4 \\ 15 & 0 & 12 \end{pmatrix} 
\]

## Ex 2.
Given matrices: 

$$
A = \begin{bmatrix} 
2 & -1\\ 
0 & 4\\ 
3 & 5 
\end{bmatrix}, \quad
B = \begin{bmatrix} 
1 & 3 & -2 & 5 \\ 
4 & -1 & 0 & 9
\end{bmatrix}
$$

- Determine the dimensions of $A$ and $B$.
- Compute $AB$.
- Is it possible to compute $BA$? Why?

## Matrix Calculations

Given matrices: 

\[ 
A = \begin{bmatrix} 
2 & -1\\ 
0 & 4\\ 
3 & 5 
\end{bmatrix}, \quad
B = \begin{bmatrix} 
1 & 3 & -2 & 5 \\ 
4 & -1 & 0 & 9
\end{bmatrix}
\]

### Dimensions of Matrices
- The dimension of matrix \( A \) is \( 3 \times 2 \).
- The dimension of matrix \( B \) is \( 2 \times 4 \).

### Matrix Multiplication \( AB \)
Since \( A \) is a \( 3 \times 2 \) matrix and \( B \) is a \( 2 \times 4 \) matrix, the product \( AB \) is defined and will result in a \( 3 \times 4 \) matrix.

\[ 
AB = \begin{bmatrix} 
2 & -1\\ 
0 & 4\\ 
3 & 5 
\end{bmatrix} 
\begin{bmatrix} 
1 & 3 & -2 & 5 \\ 
4 & -1 & 0 & 9
\end{bmatrix} 
= \begin{bmatrix} 
2 \cdot 1 + (-1) \cdot 4 & 2 \cdot 3 + (-1) \cdot (-1) & 2 \cdot (-2) + (-1) \cdot 0 & 2 \cdot 5 + (-1) \cdot 9 \\ 
0 \cdot 1 + 4 \cdot 4 & 0 \cdot 3 + 4 \cdot (-1) & 0 \cdot (-2) + 4 \cdot 0 & 0 \cdot 5 + 4 \cdot 9 \\ 
3 \cdot 1 + 5 \cdot 4 & 3 \cdot 3 + 5 \cdot (-1) & 3 \cdot (-2) + 5 \cdot 0 & 3 \cdot 5 + 5 \cdot 9 
\end{bmatrix} 
= \begin{bmatrix} 
2 - 4 & 6 + 1 & -4 + 0 & 10 - 9 \\ 
0 + 16 & 0 - 4 & 0 + 0 & 0 + 36 \\ 
3 + 20 & 9 - 5 & -6 + 0 & 15 + 45 
\end{bmatrix} 
= \begin{bmatrix} 
-2 & 7 & -4 & 1 \\ 
16 & -4 & 0 & 36 \\ 
23 & 4 & -6 & 60 
\end{bmatrix} 
\]

### Possibility of Computing \( BA \)
To compute \( BA \), \( B \) must have as many columns as \( A \) has rows. However, \( B \) is a \( 2 \times 4 \) matrix and \( A \) is a \( 3 \times 2 \) matrix. The number of columns in \( B \) (4) does not match the number of rows in \( A \) (3). Therefore, the product \( BA \) is not defined.

## Ex 3.

Let:

$$
A = \begin{bmatrix} 
1 & 2\\ 
3 & 4
\end{bmatrix}, \quad
B = \begin{bmatrix} 
0 & 1\\ 
1 & 0
\end{bmatrix}
$$

- Compute $AB$
- Compute $BA$
- Does $AB$ = $BA$? Explain why.

## Matrix Multiplication

Given matrices: 

\[ 
A = \begin{bmatrix} 
1 & 2\\ 
3 & 4
\end{bmatrix}, \quad
B = \begin{bmatrix} 
0 & 1\\ 
1 & 0
\end{bmatrix}
\]

### Matrix Multiplication \( AB \)

\[ 
AB = \begin{bmatrix} 
1 & 2\\ 
3 & 4
\end{bmatrix} 
\begin{bmatrix} 
0 & 1\\ 
1 & 0
\end{bmatrix} 
= \begin{bmatrix} 
1 \cdot 0 + 2 \cdot 1 & 1 \cdot 1 + 2 \cdot 0 \\ 
3 \cdot 0 + 4 \cdot 1 & 3 \cdot 1 + 4 \cdot 0 
\end{bmatrix} 
= \begin{bmatrix} 
2 & 1 \\ 
4 & 3 
\end{bmatrix} 
\]

### Matrix Multiplication \( BA \)

\[ 
BA = \begin{bmatrix} 
0 & 1\\ 
1 & 0
\end{bmatrix} 
\begin{bmatrix} 
1 & 2\\ 
3 & 4
\end{bmatrix} 
= \begin{bmatrix} 
0 \cdot 1 + 1 \cdot 3 & 0 \cdot 2 + 1 \cdot 4 \\ 
1 \cdot 1 + 0 \cdot 3 & 1 \cdot 2 + 0 \cdot 4 
\end{bmatrix} 
= \begin{bmatrix} 
3 & 4 \\ 
1 & 2 
\end{bmatrix} 
\]

### Comparison of \( AB \) and \( BA \)

\[ 
AB = \begin{bmatrix} 
2 & 1 \\ 
4 & 3 
\end{bmatrix} \neq BA = \begin{bmatrix} 
3 & 4 \\ 
1 & 2 
\end{bmatrix} 
\]

### Explanation

The multiplication of matrices is generally not commutative, which means \( AB \) does not necessarily equal \( BA \). In this case, \( AB \) and \( BA \) are different matrices, illustrating that matrix multiplication depends on the order of the factors.

## Ex 4.

## Matrix Inverse

Given the matrix \( A \):

\[ 
A = \begin{bmatrix} 
2 & 1 & 3 \\ 
1 & 4 & 2 \\
3 & 5 & 1
\end{bmatrix}
\]

### Finding the Inverse of Matrix \( A \)

To find the inverse of matrix \( A \), we use the formula \( A^{-1} = \frac{1}{\det(A)} \text{adj}(A) \), where \(\det(A)\) is the determinant of \( A \) and \(\text{adj}(A)\) is the adjugate of \( A \).

1. Calculate the determinant of \( A \):

\[ 
\det(A) = 2 \begin{vmatrix} 4 & 2 \\ 5 & 1 \end{vmatrix} - 1 \begin{vmatrix} 1 & 2 \\ 3 & 1 \end{vmatrix} + 3 \begin{vmatrix} 1 & 4 \\ 3 & 5 \end{vmatrix}
\]

\[ 
\det(A) = 2 (4 \cdot 1 - 2 \cdot 5) - 1 (1 \cdot 1 - 2 \cdot 3) + 3 (1 \cdot 5 - 4 \cdot 3)
\]

\[ 
\det(A) = 2 (4 - 10) - 1 (1 - 6) + 3 (5 - 12)
\]

\[ 
\det(A) = 2 (-6) - 1 (-5) + 3 (-7)
\]

\[ 
\det(A) = -12 + 5 - 21 = -28
\]

Since \(\det(A) \neq 0\), the matrix \( A \) is invertible.

2. Calculate the adjugate of \( A \):

\[ 
\text{adj}(A) = \begin{bmatrix} 
\begin{vmatrix} 4 & 2 \\ 5 & 1 \end{vmatrix} & -\begin{vmatrix} 1 & 2 \\ 5 & 1 \end{vmatrix} & \begin{vmatrix} 1 & 4 \\ 4 & 2 \end{vmatrix} \\ 
-\begin{vmatrix} 1 & 3 \\ 3 & 1 \end{vmatrix} & \begin{vmatrix} 2 & 3 \\ 3 & 1 \end{vmatrix} & -\begin{vmatrix} 2 & 1 \\ 3 & 5 \end{vmatrix} \\
\begin{vmatrix} 1 & 3 \\ 4 & 2 \end{vmatrix} & -\begin{vmatrix} 2 & 3 \\ 1 & 4 \end{vmatrix} & \begin{vmatrix} 2 & 1 \\ 1 & 4 \end{vmatrix}
\end{bmatrix}
\]

\[ 
\text{adj}(A) = \begin{bmatrix} 
4 \cdot 1 - 2 \cdot 5 & -(1 \cdot 1 - 2 \cdot 5) & 1 \cdot 2 - 4 \cdot 4 \\ 
-(1 \cdot 1 - 3 \cdot 3) & 2 \cdot 1 - 3 \cdot 3 & -(2 \cdot 5 - 1 \cdot 3) \\
1 \cdot 2 - 3 \cdot 4 & -(2 \cdot 4 - 3 \cdot 1) & 2 \cdot 4 - 1 \cdot 1 
\end{bmatrix}
\]

\[ 
\text{adj}(A) = \begin{bmatrix} 
4 - 10 & -(1 - 10) & 2 - 16 \\ 
-(1 - 9) & 2 - 9 & -(10 - 3) \\
2 - 12 & -(8 - 3) & 8 - 1 
\end{bmatrix}
\]

\[ 
\text{adj}(A) = \begin{bmatrix} 
-6 & 9 & -14 \\ 
-8 & -7 & -7 \\
-10 & -5 & 7 
\end{bmatrix}
\]

3. Calculate the inverse of \( A \):

\[ 
A^{-1} = \frac{1}{-28} \begin{bmatrix} 
-6 & 9 & -14 \\ 
-8 & -7 & -7 \\
-10 & -5 & 7 
\end{bmatrix}
\]

\[ 
A^{-1} = \begin{bmatrix} 
\frac{3}{14} & -\frac{9}{28} & \frac{7}{14} \\ 
\frac{2}{7} & \frac{1}{4} & \frac{1}{4} \\
\frac{5}{14} & \frac{5}{28} & -\frac{1}{4} 
\end{bmatrix}
\]

### Verification of Inverse

To verify that multiplying \( A \) with its inverse results in the identity matrix, we compute \( AA^{-1} \):

\[ 
AA^{-1} = \begin{bmatrix} 
2 & 1 & 3 \\ 
1 & 4 & 2 \\

\end{bmatrix} 
\begin{bmatrix}

\[

1 \cdot \frac{3}{14} + 4 \cdot \frac{2}{7} + 2 \cdot \frac{5}{14} & 1 \cdot -\frac{9}{28} + 4 \cdot \frac{1}{4} + 2 \cdot \frac{5}{28} & 1 \cdot \frac{7}{14} + 4 \cdot \frac{1}{4} + 2 \cdot -\frac{1}{4} \\ 
3 \cdot \frac{3}{14} + 5 \cdot \frac{2}{7} + 1 \cdot \frac{5}{14} & 3 \cdot -\frac{9}{28} + 5 \cdot \frac{1}{4} + 1 \cdot \frac{5}{28} & 3 \cdot \frac{7}{14} + 5 \cdot \frac{1}{4} + 1 \cdot -\frac{1}{4}

## Ex 5.
In linear regression, the solution to $Xw = y$ is given by:
$$w = (X^T X)^{-1} X^T y$$

Given:
$$X = \begin{bmatrix} 1 & 2 \\ 3 & 4 \\ 5 & 6 \end{bmatrix}, \quad y = \begin{bmatrix} 7 \\ 8 \\ 9 \end{bmatrix}$$

- Find $w$

## Linear Regression Solution

In linear regression, the solution to \( Xw = y \) is given by:
\[ w = (X^T X)^{-1} X^T y \]

Given:
\[ 
X = \begin{bmatrix} 
1 & 2 \\ 
3 & 4 \\ 
5 & 6 
\end{bmatrix}, \quad
y = \begin{bmatrix} 
7 \\ 
8 \\ 
9 
\end{bmatrix}
\]

### Step-by-Step Calculation

1. Compute \( X^T \):

\[ 
X^T = \begin{bmatrix} 
1 & 3 & 5 \\ 
2 & 4 & 6 
\end{bmatrix}
\]

2. Compute \( X^T X \):

\[ 
X^T X = \begin{bmatrix} 
1 & 3 & 5 \\ 
2 & 4 & 6 
\end{bmatrix} 
\begin{bmatrix} 
1 & 2 \\ 
3 & 4 \\ 
5 & 6 
\end{bmatrix} 
= \begin{bmatrix} 
1 \cdot 1 + 3 \cdot 3 + 5 \cdot 5 & 1 \cdot 2 + 3 \cdot 4 + 5 \cdot 6 \\ 
2 \cdot 1 + 4 \cdot 3 + 6 \cdot 5 & 2 \cdot 2 + 4 \cdot 4 + 6 \cdot 6 
\end{bmatrix} 
= \begin{bmatrix} 
1 + 9 + 25 & 2 + 12 + 30 \\ 
2 + 12 + 30 & 4 + 16 + 36 
\end{bmatrix} 
= \begin{bmatrix} 
35 & 44 \\ 
44 & 56 
\end{bmatrix}
\]

3. Compute the inverse of \( X^T X \):

\[ 
(X^T X)^{-1} = \frac{1}{\det(X^T X)} \begin{bmatrix} 
56 & -44 \\ 
-44 & 35 
\end{bmatrix}
\]

Where the determinant \(\det(X^T X)\) is calculated as follows:

\[ 
\det(X^T X) = 35 \cdot 56 - 44 \cdot 44 = 1960 - 1936 = 24
\]

Thus:

\[ 
(X^T X)^{-1} = \frac{1}{24} \begin{bmatrix} 
56 & -44 \\ 
-44 & 35 
\end{bmatrix} 
= \begin{bmatrix} 
\frac{56}{24} & -\frac{44}{24} \\ 
-\frac{44}{24} & \frac{35}{24} 
\end{bmatrix} 
= \begin{bmatrix} 
\frac{7}{3} & -\frac{11}{6} \\ 
-\frac{11}{6} & \frac{35}{24} 
\end{bmatrix}
\]

4. Compute \( X^T y \):

\[ 
X^T y = \begin{bmatrix} 
1 & 3 & 5 \\ 
2 & 4 & 6 
\end{bmatrix} 
\begin{bmatrix} 
7 \\ 
8 \\ 
9 
\end{bmatrix} 
= \begin{bmatrix} 
1 \cdot 7 + 3 \cdot 8 + 5 \cdot 9 \\ 
2 \cdot 7 + 4 \cdot 8 + 6 \cdot 9 
\end{bmatrix} 
= \begin{bmatrix} 
7 + 24 + 45 \\ 
14 + 32 + 54 
\end{bmatrix} 
= \begin{bmatrix} 
76 \\ 
100 
\end{bmatrix}
\]

5. Finally, compute \( w \):

\[ 
w = (X^T X)^{-1} X^T y = \begin{bmatrix} 
\frac{7}{3} & -\frac{11}{6} \\ 
-\frac{11}{6} & \frac{35}{24} 
\end{bmatrix} 
\begin{bmatrix} 
76 \\ 
100 
\end{bmatrix}
\]

\[ 
w = \begin{bmatrix} 
\frac{7}{3} \cdot 76 + -\frac{11}{6} \cdot 100 \\ 
-\\frac{11}{6} \cdot 76 + \frac{35}{24} \cdot 100 
\end{bmatrix}
\]

\[ 
= \begin{bmatrix} 
\frac{532}{3} - \frac{1100}{6} \\ 
-\frac{836}{6} + \frac{3500}{24} 
\end{bmatrix}
\]

\[ 
= \begin{bmatrix} 
\frac{532}{3} - \frac{550}{3} \\ 
-\frac{418}{3} + \frac{875}{6} 
\end{bmatrix}
\]

\[ 
= \begin{bmatrix} 
\frac{532 - 550}{3} \\ 
-\frac{836}{6} + \frac{875}{6}
\end{bmatrix}
\]

\[ 
= \begin{bmatrix} 
-\frac{18}{3} \\ 
\frac{39}{6}
\end{bmatrix}
\]

\[ 
= \begin{bmatrix} 
-6 \\ 
\frac{13}{2}
\end{bmatrix}
\]

So the solution vector \( w \) is:

\[ 
w = \begin{bmatrix} 
-6 \\ 
\frac{13}{2}
\end{bmatrix}
\]
