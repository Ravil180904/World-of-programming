Task 2 numerical solution for (3.1):

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda_ = 0.1 # Example value for lambda
#T0 = 10^4
T0=10^2 # Example value for T0
t_in = 10^3  # Initial condition T(0)
dt = 0.01  # Time step size
t_end = 1000  # Total number of time steps

# Initialize array to store T values
T = np.zeros(t_end)
T[0] = t_in

# Finite Difference Method for Equation (3.1)
for i in range(0, t_end-1):
    T[i+1] = T[i] + dt * lambda_ * T[i] * (1 - T[i] / T0)

# Plot the results
time = np.arange(0, t_end * dt, dt)
plt.figure(figsize=(8,6))
plt.plot(time, T)
plt.xlabel('Time')
plt.ylabel('Tumor Cell Density T(t)')
plt.title('Tumor Cell Proliferation without Immune Response')
#plt.legend()
plt.grid()
plt.show()

```
